# Linker Constructor Notebook

This notebook call linker's token constructor service functionality including:
- Clearing databases
- Constructing tokens from word special file
- Constructing tokens from code jsonl dataset
- Processing raw content directly
- Monitoring token count

```python
from src.nano import Nano
from datasets import load_dataset
```

```python
nano = Nano()
```

```python
# construct word special index
nano.construct_word_special("../dataset/tokenizer/word_*.jsonl")
print(f"Token number: {nano.get_num()}")
```

```python
# construct code jsonl dataset
nano.construct_from_glob("../download/code/go_connec*.jsonl")
print(f"Token number: {nano.get_num()}")
```

```python
ds = load_dataset("nampdn-ai/tiny-webtext")
ds
```

```python
# * process training set
for i in range(len(ds['train'])):
    content = ds['train'][i]['bot']
    filename = f"nampdn-tiny-webtext/train:{i}"
    nano.construct_content(filename, content)

    if (i + 1) % 1000 == 0:
        print(f"Processed {i+1}/{len(ds['train'])} entries. Token count: {nano.get_num()}")

print(f"Final token count: {nano.get_num()}")
```
