```python
import torch
import torch.nn as nn
```

```python
# Step 1: Create embedding table
vocab_size = 65
embedding_dim = 768
embedding_layer = nn.Embedding(vocab_size, embedding_dim)

# Step 2: Look at the weight matrix (initially random)
print("Embedding table shape:", embedding_layer.weight.shape)
# Output: torch.Size([65, 768])

print("First token's embedding (token ID 0):")
print(embedding_layer.weight[0])
# Output: tensor([-0.1234, 0.5678, -0.9012, ..., 0.3456], requires_grad=True)

# Step 3: Convert token IDs to embeddings
token_ids = torch.tensor([20, 43, 50, 50, 51])
embeddings = embedding_layer(token_ids)

print("Token IDs:", token_ids)
# Output: tensor([20, 43, 50, 50, 51])

print("Embeddings shape:", embeddings.shape)
# Output: torch.Size([5, 768])

print("Embedding for token 20 (first token 'H'):")
print(embeddings[0])
# Output: tensor([-0.2345, 0.6789, 0.1234, ..., 0.8901], grad_fn=<SelectBackward>)

# Verify it's the same as directly indexing the weight matrix:
print("Are they the same?", torch.equal(embeddings[0], embedding_layer.weight[20]))
# Output: True
```
